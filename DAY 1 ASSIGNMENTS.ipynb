```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#QUESTION 1 

x=np.arange(40,50)
y=np.arange(50,60)
plt.plot(x,y,'go--',linewidth=2,markersize=7)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#QUESTION 2

days = [1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180]
sales_2 = [70,90,160,150,140,145,175]
plt.title("Sales Trend")
plt.xlabel("Days of week")
plt.ylabel("Sales")
plt.plot(days, sales_1, linestyle='dashed', color='red', marker ='o')
plt.plot(days, sales_2, linestyle='dashed', color='green', marker ='o')
plt.grid()
plt.show()
```
